Plotting multiple charts in a grid

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Example data
years = [2000, 2001, 2002, 2003, 2004]
oranges = [10, 15, 7, 10, 12]
apples = [5, 8, 6, 9, 10]  # Make sure this is defined

# Create a sample flowers DataFrame for demonstration
data = {
    'sepal_length': [5.1, 4.9, 4.7, 4.6, 5.0],
    'sepal_width': [3.5, 3.0, 3.2, 3.1, 3.6],
    'species': ['setosa', 'setosa', 'versicolor', 'versicolor', 'virginica']
}
flowers_df = pd.DataFrame(data)

# Create separate DataFrames for each species
setosa_df = flowers_df[flowers_df['species'] == 'setosa']
versicolor_df = flowers_df[flowers_df['species'] == 'versicolor']
virginica_df = flowers_df[flowers_df['species'] == 'virginica']

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(12, 9))

# Plot on the first subplot
axes[0, 0].plot(years, apples, 's--b', label='Apples')  
axes[0, 0].plot(years, oranges, 'o--r', label='Oranges')
axes[0, 0].set_xlabel('Year')  
axes[0, 0].set_ylabel('Yield (tons per hectare)')  
axes[0, 0].set_title('Yield of Oranges and Apples')
axes[0, 0].legend()

# Plot on the second subplot
axes[0, 1].set_title('Sepal Length vs. Width')
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', s=100, data=flowers_df, ax=axes[0, 1])

# Plot on the third subplot: Distribution of Sepal Width
axes[0, 2].set_title('Distribution of Sepal Width')
axes[0, 2].hist([setosa_df['sepal_width'], versicolor_df['sepal_width'], virginica_df['sepal_width']], 
                bins=np.arange(2, 5, 0.25), stacked=True)
axes[0, 2].legend(['Setosa', 'Versicolor', 'Virginica'])

# Example DataFrame for restaurant bills
tips_data = {
    'day': ['Thur', 'Fri', 'Sat', 'Sun'],
    'total bill': [10, 20, 30, 40],
    'sex': ['Female', 'Male', 'Female', 'Male']
}
tips_df = pd.DataFrame(tips_data)

# Pass the axes into seaborn for the fourth subplot
axes[1, 0].set_title('Restaurant Bills')
sns.barplot(x='day', y='total bill', hue='sex', data=tips_df, ax=axes[1, 0])

# Example data for flight traffic
flights_data = np.random.rand(10, 12)  # Dummy data for heatmap
flights_df = pd.DataFrame(flights_data)

# Pass the axes into seaborn for the fifth subplot
axes[1, 1].set_title('Flight Traffic')
sns.heatmap(flights_df, cmap='Blues', ax=axes[1, 1])

# Load the image for the sixth subplot
axes[1, 2].set_title('A Data Science Pic')
img = plt.imread(r'C:\Users\CHARAN RAJ JKS\Pictures\IMG_20230828_120552.jpg')
axes[1, 2].imshow(img)
axes[1, 2].grid(False)
axes[1, 2].set_xticks([])
axes[1, 2].set_yticks([])

plt.tight_layout(pad=2)
plt.show()

```
